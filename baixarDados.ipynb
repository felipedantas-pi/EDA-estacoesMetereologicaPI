```python
import requests
import pandas as pd
```

```python
url_link = 'https://apimapas.inmet.gov.br/estacoes'
header = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36',
}
```

```python
response = requests.get(url_link, headers=header).json()
```

```python
response
```

```python
type(response)
```

```python
df_regiao = pd.json_normalize(response, max_level=2)
df_regiao
```

```python
df_ne = df_regiao['estacoes.automaticas.NE'].explode().apply(pd.Series)
df_ne
```

```python
df_ne.info()
```
