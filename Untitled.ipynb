## Recuperando dados da série histórica da API HidroWeb

A seguinte URL busca os dados das estacoes pelo código da mesma.

```python
requests.get(f"https://www.snirh.gov.br/hidroweb/rest/api/estacaotelemetrica?id={241000}").json()['content']
```

```python
dados_estacoes = []
for cod in cod_list:
    print("Iniciando a interação")
    print(f"O código é: {cod}")
    r = requests.get(f"https://www.snirh.gov.br/hidroweb/rest/api/estacaotelemetrica?id={cod}").json()['content']
    if len(r) > 0:
        dados_estacoes.append(r)
        print("Dados adicionados")
    print(r)
```
